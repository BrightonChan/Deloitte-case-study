# Overall Sales Analysis for Soda

```python
monthly_sales = soda.groupby('month')['total_price'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['month'], monthly_sales['total_price'], marker='o', linestyle='-', color='blue')
plt.title('Total Sales by Month (Aggregated Over Years)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(monthly_sales['month'], labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
yearly_monthly_sales = soda.groupby(['year', 'quarter'])['total_price'].sum().reset_index()

# Plotting
plt.figure(figsize=(14, 8))
for year in yearly_monthly_sales['year'].unique():
    data = yearly_monthly_sales[yearly_monthly_sales['year'] == year]
    plt.plot(data['quarter'], data['total_price'], marker='o', linestyle='-', label=f'year {year}')

plt.title('Quarterly Sales Trends by year')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(range(0, 4), labels=['Q1', 'Q2', 'Q3', 'Q4'])
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
heatmap_data = yearly_monthly_sales.pivot_table(index='quarter', columns='year', values='total_price', fill_value=0)


plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)
plt.title('Heatmap of Quarterly Sales by Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.yticks(range(4), labels=['Q1', 'Q2', 'Q3', 'Q4'], rotation=0)
plt.show()
```

# Sales Analysis between Red Bull and Monster

```python
bev = sales[sales["desc"] == "Beverage - Energy/Protein"]
bev
```

```python
bev['quarter_int'] = bev['quarter'].str.extract('(\d)').astype(int)
bev['quarter_int']
```

```python
bev['year_quarter'] = pd.PeriodIndex(year=bev['year'], quarter=bev['quarter_int'], freq='Q')
bev
```

```python
red_bull = bev[bev["item_name"].str.contains("Red Bull")]
red_bull.head()
```

```python
monster = bev[bev["item_name"].str.contains("Monster")]
monster.head()
```

```python
plt.figure(figsize=(12, 6))

# Plotting Red Bull sales
red_bull_sales = red_bull.groupby('month')['total_price'].sum()
plt.plot(red_bull_sales.index, red_bull_sales.values, label='Red Bull')

# Plotting Monster sales
monster_sales = monster.groupby('month')['total_price'].sum()
plt.plot(monster_sales.index, monster_sales.values, label='Monster')

# Adding titles and labels
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Plotting Red Bull sales
red_bull_sales = red_bull.groupby('year')['total_price'].sum()[:-1]
plt.plot(red_bull_sales.index, red_bull_sales.values, label='Red Bull')

# Plotting Monster sales
monster_sales = monster.groupby('year')['total_price'].sum()[:-1]
plt.plot(monster_sales.index, monster_sales.values, label='Monster')

# Adding titles and labels
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend()

# Display the plot
plt.show()
```

```python
red_bull_sales = red_bull.groupby('year_quarter')['total_price'].sum()[:-1]
red_bull_sales.index = red_bull_sales.index.to_timestamp()

# Prepare Monster data
monster_sales = monster.groupby('year_quarter')['total_price'].sum()[:-1]
monster_sales.index = monster_sales.index.to_timestamp()

plt.figure(figsize=(12, 6))

# Define colormap
colors = plt.cm.tab10(np.linspace(0, 1, 4))

# Plot Red Bull sales line and scatter
plt.plot(red_bull_sales.index, red_bull_sales, marker='', color='blue', linewidth=1, label='Red Bull')
for i, quarter in enumerate(range(1, 5)):
    quarter_sales = red_bull_sales[red_bull_sales.index.quarter == quarter]
    plt.scatter(quarter_sales.index, quarter_sales, color=colors[i], label=f'Red Bull Q{quarter}')

# Plot Monster sales line and scatter
plt.plot(monster_sales.index, monster_sales, marker='', color='green', linewidth=1, label='Monster')
for i, quarter in enumerate(range(1, 5)):
    quarter_sales = monster_sales[monster_sales.index.quarter == quarter]
    plt.scatter(quarter_sales.index, quarter_sales, color=colors[i], marker='x', label=f'Monster Q{quarter}')

# Set custom x-ticks
plt.xticks(ticks=red_bull_sales.index, labels=red_bull_sales.index.to_period('Q').astype(str), rotation=45)

# Adding title and labels
plt.title('Total Sales by Quarter for Red Bull and Monster')
plt.xlabel('Year-Quarter')
plt.ylabel('Total Sales')
plt.legend(title='Brand and Quarter')

plt.show()
```

# Sales analysis between different types of coke

```python
coke_products = [
    'Diet Coke - 12 oz cans', 'Coke Classic 12 oz cans', 'Coke Zero Sugar 12 oz cans', 
    'Cherry Coke 12oz', 'Cherry Coke Zero 12 pack', 'Diet Coke Caffeine Free 12oz cans'
]

coke_data = soda[soda['item_name'].isin(coke_products)]
```

```python
coke_sales = coke_data.groupby('item_name')['total_price'].sum().reset_index()
```

```python
coke_monthly_sales = coke_data.groupby(['item_name', 'year', 'month'])['total_price'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(coke_sales['item_name'], coke_sales['total_price'], color='skyblue')
plt.title('Total Sales by Coke Product')
plt.xlabel('Coke Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 8))

# Plot each Coke product's monthly sales
for product in coke_sales['item_name']:
    product_data = coke_monthly_sales[coke_monthly_sales['item_name'] == product]
    plt.plot(product_data['month'], product_data['total_price'], marker='o', linestyle='-', label=product)

plt.title('Monthly Sales Trends for Coke Products')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Coke Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
total_coke_sales = coke_sales['total_price'].sum()
coke_sales['market_share'] = (coke_sales['total_price'] / total_coke_sales) * 100

print(coke_sales)

# Plot market share by Coke product
plt.figure(figsize=(12, 6))
plt.pie(coke_sales['market_share'], labels=coke_sales['item_name'], autopct='%1.1f%%', startangle=140)
plt.title('Market Share of Coke Products')
plt.show()
```

```python
heatmap_data = coke_monthly_sales.pivot_table(index='month', columns='item_name', values='total_price', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)
plt.title('Heatmap of Monthly Sales for Coke Products')
plt.xlabel('Coke Product')
plt.ylabel('Month')
plt.yticks(range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
plt.show()
```
